```python
import cv2
from matplotlib import pyplot as plt

#Load image and convert color domain
path = 'images/beach.png'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Calculate equalized histogram
eq = cv2.equalizeHist(image)

#Calculate histogram
Ohist = cv2.calcHist([image], [0], None, [256], [0, 256])
Ehist = cv2.calcHist([eq], [0], None, [256], [0, 256])

#Plot image
fig1 = plt.figure(figsize=(20, 12))
ax1 = fig1.add_subplot(2,2,1)
ax1.set(title='Original gray image')
ax1.imshow(image, cmap='gray')
#######################################
ax2 = fig1.add_subplot(2,2,2)
ax2.set(title='Equalized gray image')
ax2.imshow(eq, cmap='gray')
#######################################
ax3 = fig1.add_subplot(2,2,3)
ax3.set(title='Histogram of original image')
ax3.set_xlabel('Bins of intensity')
ax3.set_ylabel('# of pixels')
ax3.plot(Ohist, color='black')
#######################################
ax4 = fig1.add_subplot(2,2,4)
ax4.set(title='Histogram of Equalized image')
ax4.set_xlabel('Bins of intensity')
ax4.set_ylabel('# of pixels')
ax4.plot(Ehist, color='black')
```
