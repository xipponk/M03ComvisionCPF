```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Create array for bitwise operation
Rectangle = np.zeros((300,300), dtype='uint8')
Circle = np.zeros((300,300), dtype='uint8')

#Drawing rectangle and circle to our array
cv2.rectangle(Rectangle, (25,25), (275,275), 255, -1)
cv2.circle(Circle, (150,150), 150, 255, -1)

#Plot image
fig1 = plt.figure(figsize=(8, 5))
ax1 = fig1.add_subplot(1,2,1)
ax1.set(title='Rectangle')
ax1.imshow(Rectangle, cmap='gray')
ax2 = fig1.add_subplot(1,2,2)
ax2.set(title='Circle')
ax2.imshow(Circle, cmap='gray')

#Bitwise Operation
AND = cv2.bitwise_and(Rectangle, Circle)
OR = cv2.bitwise_or(Rectangle, Circle)
XOR = cv2.bitwise_xor(Rectangle, Circle)
NOT_Rectangle = cv2.bitwise_not(Rectangle)
NOT_Circle = cv2.bitwise_not(Circle)

#Plot image
fig2 = plt.figure(figsize=(16, 10))
ax1 = fig2.add_subplot(2,3,1)
ax1.set(title='AND')
ax1.imshow(AND, cmap='Pastel1')
ax2 = fig2.add_subplot(2,3,2)
ax2.set(title='OR')
ax2.imshow(OR, cmap='Pastel1')
ax3 = fig2.add_subplot(2,3,3)
ax3.set(title='XOR')
ax3.imshow(XOR, cmap='gray')
ax4 = fig2.add_subplot(2,3,4)
ax4.set(title='NOT Rectangle')
ax4.imshow(NOT_Rectangle, cmap='Pastel1')
ax5 = fig2.add_subplot(2,3,5)
ax5.set(title='NOT Circle')
ax5.imshow(NOT_Circle, cmap='Pastel1')
```
