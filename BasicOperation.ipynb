```python
import cv2
from matplotlib import pyplot as plt

#Load image and convert color domain
image = cv2.imread('images/messi.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Plot original image
fig1 = plt.figure()
ax1 = fig1.add_subplot(1,1,1)
ax1.set(title='Original Image')
ax1.imshow(image)

#Access image properties, size and datytype
print(image.shape)
print('The image has height {} and width {} with {} channels.'.format(image.shape[0], image.shape[1], image.shape[2]))
print('The size of image is {} pixels.'.format(image.size))
print('The datatype of image is {}.'.format(image.dtype))
```

```python
#Image Region of Interest (ROI)
roi = image.copy()
ball = image[280:340, 330:390]
roi[273:333, 100:160] = ball

#Plot ROI image
fig2 = plt.figure()
ax1 = fig2.add_subplot(1,1,1)
ax1.set(title='ROI Image')
ax1.imshow(roi)
```

```python
#Split and merge image
r, g, b = cv2.split(image)

#Plot image in R, G, B channels
fig3 = plt.figure(figsize=(15, 8))
ax1 = fig3.add_subplot(1,3,1)
ax1.set(title='R Channel')
ax1.imshow(r)
ax2 = fig3.add_subplot(1,3,2)
ax2.set(title='G Channel')
ax2.imshow(g, cmap='gray')
ax3 = fig3.add_subplot(1,3,3)
ax3.set(title='B Channel')
ax3.imshow(b, cmap='gray')

merged = cv2.merge((r,g,b))
fig4 = plt.figure()
ax1 = fig4.add_subplot(1,1,1)
ax1.set(title='Merged Image')
ax1.imshow(merged)
```

```python
image2 = image.copy()

#fill 0 in channel red
image2[:,:,0] = 0

r2 = image2[:,:,0]
g2 = image2[:,:,1]
b2 = image2[:,:,2]

merged2 = cv2.merge((r2, g2, b2))
fig5 = plt.figure()
ax1 = fig5.add_subplot(1,1,1)
ax1.set(title='Merged Image')
ax1.imshow(merged2)
```
