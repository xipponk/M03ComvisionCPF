```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Open image from path
path = 'images/trex.png'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Apply blur filters
blur1 = np.hstack([
    cv2.blur(image, (3,3)),
    cv2.blur(image, (5,5)),
    cv2.blur(image, (7,7))
])

blur2 = np.hstack([
    cv2.GaussianBlur(image,(3,3),0),
    cv2.GaussianBlur(image,(5,5),0),
    cv2.GaussianBlur(image,(7,7),0)
])

blur3 = np.hstack([
    cv2.medianBlur(image,3),
    cv2.medianBlur(image,5),
    cv2.medianBlur(image,7)
])

blur4 = np.hstack([
    cv2.bilateralFilter(image,5,20,20),
    cv2.bilateralFilter(image,7,30,30),
    cv2.bilateralFilter(image,8,40,40)
])


#Plot image
fig = plt.figure(figsize=(16, 20))
##################################
ax1 = fig.add_subplot(4,1,1)
ax1.set(title='Box or Average Blur')
ax1.imshow(blur1)
################################
ax2 = fig.add_subplot(4,1,2)
ax2.set(title='Gaussian Blur')
ax2.imshow(blur2)
################################
ax3 = fig.add_subplot(4,1,3)
ax3.set(title='Median Blur')
ax3.imshow(blur3)
################################
ax4 = fig.add_subplot(4,1,4)
ax4.set(title='Bilateral Blur')
ax4.imshow(blur4)
```
