```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Opean and Load image
path = r'images/coins.png'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Cretae gray image and blurred it!
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
blurred = cv2.GaussianBlur(gray,(5,5),0)
```

```python
#Sobel Operation with Convolution technic
kernelX = np.array([[-1, 0, 1],[-2, 0, 2], [-1, 0, 1]], dtype=np.float32)
kernelY = np.array([[1, 2, 1],[0, 0, 0], [-1, -2, -1]], dtype=np.float32)

#Convoluted kernel with image
sobelX = cv2.filter2D(blurred, cv2.CV_32F, kernelX)
sobelY = cv2.filter2D(blurred, cv2.CV_32F, kernelY)

#Take an absolute and convert to unsigned integer 8-bit
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

#Combined result of Sobel X and Y
sobelCombined = cv2.bitwise_or(sobelX, sobelY)

#Plot image
fig = plt.figure(figsize=(25, 15))
##############ax1###################
ax1 = fig.add_subplot(1,3,1)
ax1.set(title='Edge in X-direction')
ax1.imshow(sobelX, cmap='gray')
##############ax2###################
ax2 = fig.add_subplot(1,3,2)
ax2.set(title='Edge in Y-direction')
ax2.imshow(sobelY, cmap='gray')
##############ax3###################
ax3 = fig.add_subplot(1,3,3)
ax3.set(title='Total edge of X and Y')
ax3.imshow(sobelCombined, cmap='gray')
```

```python
#Soble operation with OpenCV technic
sobelX2 = cv2.Sobel(blurred, cv2.CV_32F, 1, 0, ksize=3)
sobelY2 = cv2.Sobel(blurred, cv2.CV_32F, 0, 1, ksize=3)

#Take an absolute and convert to unsigned integer 8-bit
sobelX2 = np.uint8(np.absolute(sobelX2))
sobelY2 = np.uint8(np.absolute(sobelY2))

#Combined result of Sobel X and Y
sobelCombined2 = cv2.bitwise_or(sobelX2, sobelY2)

#Plot image
fig2 = plt.figure(figsize=(25, 15))
##############ax1###################
ax1 = fig2.add_subplot(1,3,1)
ax1.set(title='(OpenCV) Edge in X-direction')
ax1.imshow(sobelX2, cmap='gray')
##############ax2###################
ax2 = fig2.add_subplot(1,3,2)
ax2.set(title='(OpenCV) Edge in Y-direction')
ax2.imshow(sobelY2, cmap='gray')
##############ax3###################
ax3 = fig2.add_subplot(1,3,3)
ax3.set(title='(OpenCV) Total edge of X and Y')
ax3.imshow(sobelCombined2, cmap='gray')
```
