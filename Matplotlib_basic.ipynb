```python
import cv2
from matplotlib import pyplot as plt

imageBGR = cv2.imread('images/messi.jpg')
imageRGB = cv2.cvtColor(imageBGR, cv2.COLOR_BGR2RGB)

#Open image with OpenCV
cv2.startWindowThread()
cv2.imshow("BGR", imageBGR)
cv2.imshow("RGB", imageRGB)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Open image with Matplotlib
plt.figure()
plt.imshow(imageRGB)
plt.title('RGB Image')
plt.xlabel('width')
plt.ylabel('height')

plt.figure()
plt.imshow(imageBGR)
plt.title('BGR Image')
plt.xlabel('width')
plt.ylabel('height')
plt.show()
```

```python
#Matplotlib subplot
plt.figure(figsize=(11, 3))
plt.subplot(1, 2, 1)
plt.imshow(imageRGB)
plt.title('RGB Image')
plt.xlabel('width')
plt.ylabel('height')
plt.subplot(1, 2, 2)
plt.imshow(imageBGR)
plt.title('BGR Image')
plt.xlabel('width')
plt.ylabel('height')
```

```python
#Matplotlib subplot (object type)
fig1 = plt.figure(figsize=(10, 3))
ax1 = fig1.add_subplot(1,2,1)
ax1.set(title='RGB Image', xlabel='width', ylabel='height')
ax1.imshow(imageRGB)
ax2 = fig1.add_subplot(1,2,2)
ax2.set(title='BGR Image', xlabel='width', ylabel='height')
ax2.imshow(imageBGR)
#fig1.savefig('testimage.png')
```
