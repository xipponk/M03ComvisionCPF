```python
import cv2
from matplotlib import pyplot as plt

#Opean and Load image
path = r'images/coins.png'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Cretae gray image and blurred it!
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
blurred = cv2.GaussianBlur(gray,(5,5),0)
'''
#Adaptive Thresholding image
cv2.adaptiveThreshold(input_image,
                      maximum_value,
                      cv2.ADAPTIVE_THRESH_MEAN_C (cv2.ADAPTIVE_THRESH_GAUSSIAN_C),
                      cv2.THRESH_BINARY_INV,
                      neighborhood size (must be odd),
                      constant (integer that subtracted from the mean))
'''
#Apply adaptive thresholding
Mthresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 111, 13)
Gthresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 111, 13)

#Plot image
fig = plt.figure(figsize=(10, 15))
##############ax1###################
ax1 = fig.add_subplot(3,2,1)
ax1.set(title='Original image')
ax1.imshow(image)
##############ax2###################
ax2 = fig.add_subplot(3,2,2)
ax2.set(title='Gray scale blurred image')
ax2.imshow(blurred, cmap='gray')
##############ax3###################
ax3 = fig.add_subplot(3,2,3)
ax3.set(title='Adaptive Mean Thresholding')
ax3.imshow(Mthresh, cmap='gray')
##############ax4###################
ax4 = fig.add_subplot(3,2,4)
ax4.set(title='Adaptive Gaussian Thresholding')
ax4.imshow(Gthresh, cmap='gray')
```
