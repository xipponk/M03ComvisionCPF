```python
import cv2
from matplotlib import pyplot as plt

#Open image from path
path = 'images/messi.jpg'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Create color for constant padding
white = (255, 255, 255)

#Add padding to image
#cv2.copyMakeBorder(input_image, px_top, px_bottom, px_left, px_right, cv2.BORDER_REPLICATE)
replicate = cv2.copyMakeBorder(image, 10, 10, 10, 10, cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(image, 10, 10, 10, 10, cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(image, 10, 10, 10, 10, cv2.BORDER_REFLECT_101)
constant = cv2.copyMakeBorder(image, 10, 10, 10, 10, cv2.BORDER_CONSTANT, value=white)

#Plot image
#Plot image in R, G, B channels
fig = plt.figure(figsize=(15, 7))
ax1 = fig.add_subplot(2,3,1)
ax1.set(title='Original image')
ax1.imshow(image)
ax2 = fig.add_subplot(2,3,2)
ax2.set(title='Replicate Padding')
ax2.imshow(replicate)
ax3 = fig.add_subplot(2,3,3)
ax3.set(title='Reflect Padding')
ax3.imshow(reflect)
ax4 = fig.add_subplot(2,3,4)
ax4.set(title='Reflect 101 Padding')
ax4.imshow(reflect101)
ax5 = fig.add_subplot(2,3,5)
ax5.set(title='Constant Padding')
ax5.imshow(constant)
```
