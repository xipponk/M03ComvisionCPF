```python
import cv2
from matplotlib import pyplot as plt

#Load image and convert color domain
path = 'images/beach.png'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Split image
r, g, b = cv2.split(image)

#Calculate histogram of each RGB channel
histR = cv2.calcHist([r], [0], None, [256], [0, 256])
histG = cv2.calcHist([g], [0], None, [256], [0, 256])
histB = cv2.calcHist([b], [0], None, [256], [0, 256])

#Plot image
fig1 = plt.figure(figsize=(20, 6))
ax1 = fig1.add_subplot(1,2,1)
ax1.set(title='Original image')
ax1.imshow(image)
ax2 = fig1.add_subplot(1,2,2)
ax2.set(title='RGB Histogram')
ax2.set_xlabel('Bins of intensity')
ax2.set_ylabel('# of pixels')
ax2.plot(histR, color='red')
ax2.plot(histG, color='green')
ax2.plot(histB, color='blue')
```
