```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Load image and convert color domain
path = 'images/beach.png'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Create masking
Rectangle = np.zeros(image.shape[0:2], dtype='uint8')
Circle = np.zeros(image.shape[0:2], dtype='uint8')
(cX, cY) = (image.shape[1]//2, image.shape[0]//2) #find (x,y) of a center point
cv2.rectangle(Rectangle, (cX-75, cY-75), (cX+75, cY+75), 255, -1)
cv2.circle(Circle, (cX, cY), 100, 255, -1)

#Masking to image
mRectangle = cv2.bitwise_and(image, image, mask=Rectangle)
mCircle = cv2.bitwise_and(image, image, mask=Circle)

#Plot image
fig1 = plt.figure(figsize=(20, 10))
ax1 = fig1.add_subplot(2,3,1)
ax1.set(title='Original image')
ax1.imshow(image)
ax2 = fig1.add_subplot(2,3,2)
ax2.set(title='Rectangle masking')
ax2.imshow(Rectangle, cmap='gray')
ax3 = fig1.add_subplot(2,3,3)
ax3.set(title='Circle masking')
ax3.imshow(Circle, cmap='gray')
ax4 = fig1.add_subplot(2,3,4)
ax4.set(title='Rectangle masked')
ax4.imshow(mRectangle)
ax5 = fig1.add_subplot(2,3,5)
ax5.set(title='Circle masked')
ax5.imshow(mCircle)
```
