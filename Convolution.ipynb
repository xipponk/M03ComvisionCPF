```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Open image from path
path = 'images/board.jpg'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Create kernel
kernel1 = np.ones((3,3), np.float32)/9 #blur kernel
kernel2 = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]]) #sharpen kernel
kernel3 = np.array([[1, 2, 1],
                   [2, 4, 2],
                   [1, 2, 1]])/16 #Gaussian blur

#Convolved image
cblurred = cv2.filter2D(image, -1, kernel1, cv2.BORDER_CONSTANT)
csharpen = cv2.filter2D(image, -1, kernel2, cv2.BORDER_CONSTANT)
gblurred = cv2.filter2D(image, -1, kernel3, cv2.BORDER_REPLICATE)

#Plot image
fig = plt.figure(figsize=(16, 10))
ax1 = fig.add_subplot(1,3,1)
ax1.set(title='Original image')
ax1.imshow(image)
ax2 = fig.add_subplot(1,3,2)
ax2.set(title='Convolution blurred')
ax2.imshow(cblurred)
ax3 = fig.add_subplot(1,3,3)
ax3.set(title='Convolution Sharpen')
ax3.imshow(csharpen)
#########################################
fig2 = plt.figure(figsize=(16, 10))
ax1 = fig2.add_subplot(1,1,1)
ax1.set(title='Gaussian blur')
ax1.imshow(gblurred)
```
